```python
import folium
```

```python
sg_latitude = 1.36
sg_longitude = 103.75
print('The geograpical coordinates of Singapore are', sg_latitude,',',sg_longitude,'.')

# create map of Singapore using latitude and longitude values
map_singapore = folium.Map(location=
                           [sg_latitude, sg_longitude], 
                           zoom_start=11.5, 
                           control_scale=True)
map_singapore

info = pd.DataFrame(
    {'Name': cc_list,
     'Region': cc_regions,
     'Contact Number':cc_contacts,
     'Latitude': cc_latitudes,
     'Longitude': cc_longitudes
    })
info #displays the dataframe
```

```python
# add markers to map
for index,row in info.iterrows():
    folium.Marker(
        [row.loc['Latitude'],row.loc['Longitude']],
        radius = 5,
        popup=row.loc['Name'] + ' Phone: '+ str(row.loc['Contact Number']),
        tooltip=row.loc['Name'],
        color= 'blue',
        fill=True,
        fill_opacity=0.7,
        parse_html=False).add_to(map_singapore)  
map_singapore
```
